```python
import numpy as np
import math
from scipy import linalg
```

```python
# введение аппроксимируемой функции
def f(x):
    return (math.sin(x/5.)*math.exp(x/10.) + 5*math.exp(-x/2.))
```

```python
# определение коэффициентов линейного приближения по точкам 1 и 15
x_1 = np.array([1, 15])
A = np.array([[1, x_1[0]], [1,x_1[1]]])
b = np.array([[f(x_1[0])], [f(x_1[1])]])
w_1 = linalg.solve(A,b)
print(w_1)
y_1 = w_1[0,0] + w_1[1,0]*x_1
print (y_1)
```

```python
%matplotlib inline
```

```python
from matplotlib import pylab as plt
```

```python
plt.plot (x_1,y_1)
plt.show
```

```python
x_0 = np.arange(0,16,0.1)
y_0 = np.array([f(i) for i in x_0])
```

```python
plt.plot (x_0, y_0)
plt.show
```

```python
# определение коэффициентов квадратичного приближения по точкам 1 8 15
x_2 = np.array([1, 8, 15])

A = np.array([[1, x_2[0], x_2[0]**2], [1,x_2[1], x_2[1]**2], [1, x_2[2], x_2[2]**2]])
b = np.array([f(x_2[0]), f(x_2[1]), f(x_2[2]) ])
w_2 = linalg.solve(A,b)
y_2 = w_2[0] + w_2[1]*x_2 + w_2[2]*(x_2**2)

print (y_2)
print (b)
```

```python
plt.plot (x_2,y_2)
plt.show
```

```python
# определение коэффициентов кубического приближения по точкам 1 4 10 15
x_3 = np.array([1, 4, 10, 15])

A = np.array([[1, x_3[0], x_3[0]**2, x_3[0]**3], [1, x_3[1], x_3[1]**2, x_3[1]**3], [1, x_3[2], x_3[2]**2, x_3[2]**3], [1, x_3[3], x_3[3]**2, x_3[3]**3]])
b = np.array([f(x_3[0]), f(x_3[1]), f(x_3[2]), f(x_3[3]) ])
w_3 = linalg.solve(A,b)
y_3 = w_3[0] + w_3[1]*x_3 + w_3[2]*(x_3**2) + w_3[3]*(x_3**3)

print (y_3)
print (b)
```

```python
plt.plot (x_3,y_3)
plt.show
```
