```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
import math

from scipy.stats import weibull_min
```

```python
# Задача 1
# Выберите ваше любимое непрерывное распределение 
# Сгенерируйте из него выборку объёма 1000, постройте гистограмму выборки
# нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины 

scale = 1
a = 1 

stat = sts.weibull_min(a = 1, scale = 1 ) 
sample = stat.rvs(1000) 
```

```python
a=1
math_exp = a*math.gamma(1 + 1/a)
disp = (a**2)*math.gamma(1+2/1) - (math_exp)**2  # среднеквадратическое отклонение
print({'math_exp':math_exp,'disp':disp})
```

```python
plt.hist(sample,density=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

x = np.linspace(0,5,10)
pdf = stat.pdf(x) #задаем теоретическую плотность распределения
h = plt.plot(x, stat.pdf(x))
```

```python
# Задача 2
# Ваша задача — оценить распределение выборочного среднего вашей случайной величины при разных объёмах выборок. 
# Для этого при трёх и более значениях n (например, 5, 10, 50) сгенерируйте 1000 выборок объёма n 
# постройте гистограммы распределений их выборочных средних. 
# Используя информацию о среднем и дисперсии исходного распределения (её можно без труда найти в википедии)
# посчитайте значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних.
```

```python
# 1 вариант: n = 5
# шаг 1
n = 5
mass=[] #назначаем пустой массив
for _ in range (1000): # задаем 1000 иттераций 
    sample = stat.rvs(n) # создаем выборку данного объема при распределение Вейбулла 
    mean = np.mean(sample) # определяем выборочное среднее
    mass.append(mean) # подгружаем данные в массив

x = np.linspace(0,5,100)

# шаг 2
# опеределим параметры норм распределения при кол-ве массивов = 1000 и n = 5 
# воспользуемся ЦПД

D_n = disp/(n)
standard_deviation = math.sqrt(D_n)

# зададим нормальное распределение с параметрами math_exp и standard_deviation

norm = sts.norm(math_exp, standard_deviation)
pdf = norm.pdf(x) 

# задаем гистограмму выборочных средних при n=5 и плотность описывающую этот массив нормального распределения
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.hist(mass,density=True)
h = plt.plot(x, norm.pdf(x))

```
