```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from math import sqrt
%matplotlib inline
```

```python
x=np.linspace(start=-10,stop=10,num=1000)
```

В данном исследовании мы проверим работу центральной предельной теоремы на примере случайной величины с распределением $\chi ^2$. [Ссылка на Wiki](https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D1%85%D0%B8-%D0%BA%D0%B2%D0%B0%D0%B4%D1%80%D0%B0%D1%82)
Дисперсия: $\sigma^2=2k$, где $k$ - математическое ожидаение.

```python
a=sts.chi2(5) #распределение chi^2
asd=a.rvs(1000)
plt.hist(asd, density=True, bins=30, label='samp_hist')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

x=np.linspace(0,20,1000)
pdf=a.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=1)
plt.title('Распределение $\chi^2(5)$')
plt.legend()
```

```python
arr=np.array([]) #создаем массив для средних
i=1
n=5
while i in range(1,1001): #цикл создания массива средних с 1000 элементами
    a=sts.chi2(5) #случайная величина распределения Хи-квадрат с 5 степенями свободы
    asd=a.rvs(n) # выбираем n значений из случайной величины а
    mean=sum(asd)/float(n) # рассчитываем выборочное среднее выборки из n значений
    arr=np.append(arr, mean) # добавляем среднее в массив
    i+=1
plt.hist(arr, bins=30, density=True, label='Sample mean, n={}'.format(n)) # гистограмма выборочных средних

avg=sum(arr)/len(arr) # мат. ожидание массива выборочных средних
d=0.0 
for i in range(0,len(arr)): # цикл рассчета суммы квадратов отклонений от среднего элементов массива выборочных средних
    d=d+(arr[i]-avg)**2 # накопленный итог суммы квадратов отклонения от среднего
    i+=1
st=sqrt(d/len(arr)) # рассчет среднего квад. отклонения
x=np.linspace(0,10,1000) # генерация последовательности
b=sts.norm(avg,st) # генерация случайной величины с со степенями свободы agv, st^2
pdf1=b.pdf(x) # функция плотности распределения
plt.plot(x, pdf1, label='Density function') # график функции плотности
plt.ylabel('Fraction of samples')
plt.xlabel('Sampling distribution of the mean')
plt.legend()
print('Мат. ожидание: E={}'.format(round(avg,2)))
print('Дисперсия: D={}'.format(round(st**2,2)))
```
