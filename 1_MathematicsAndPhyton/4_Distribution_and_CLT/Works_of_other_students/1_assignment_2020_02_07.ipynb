```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
# генерация выборки обьемом 1000:
rv = sts.expon()
sample = rv.rvs(1000)
```

```python
x = np.linspace(0,10,1000)
pdf = rv.pdf(x)
```

```python
plt.xlabel('$x$')
plt.ylabel('$f(x)$')

# граффик и гистограмма
plt.plot(x, pdf, linewidth = 2.5)
plt.hist(sample, density = True)
```

```python
def generate_pdf(n): 
    means = []

    # герерация выборочных средних по выборкам объема n
    for _ in range(1000):
        sample = rv.rvs(n)
        sample_mean = np.mean(sample)
        means.append(sample_mean)

    # теоритеческая плотность вероятности для нормального распределения
    mu = 1
    sigma = (1./n) ** 0.5
    norm_rv = sts.norm(loc=mu, scale=sigma)
    x = np.linspace(0, 10, 1000)
    pdf = norm_rv.pdf(x)

    # создание гистограммы выборочных средних
    plt.plot(x, pdf, linewidth = 2.5)
    plt.hist(means, density = True)
    plt.xlabel('$\\bar{X}_n$')
    plt.ylabel('$f(\\bar{X}_n)$')
    plt.title('n = %d' % n)
    plt.show()
```

```python
# созднаие гистограмм и приближающих их нормальных распределений при различных объемах выборок n
for n in [1, 5, 25, 50, 100, 500]:
    generate_pdf(n)
```

### ВЫВОДЫ:
1. чем больше объем выборки n, тем лучше распределение выборочных средних соответствует приближающему его нормальному распределению
2. с ростом n уменьшается дисперсия распределения выборочных средних
