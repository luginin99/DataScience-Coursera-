```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

# Исследую хи-квадртат расспределение с четырьмя степенями свободы
(https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D1%85%D0%B8-%D0%BA%D0%B2%D0%B0%D0%B4%D1%80%D0%B0%D1%82) 
## Гистограмма и функция плотности вероятности
Для справки: плотность хи-квадрат распределения вычисляется по формуле $$ f_{\chi^2(k)}(x) = \frac{(1/2)^\frac{k}{2}}{\Gamma(\frac{k}{2})}x^{\frac{k}{2}-1}e^{-\frac{x}{2}} $$ где k - количество степеней свободы

```python
chi2_rv = sts.chi2(4)
# График теоретической функции распределения
x = np.linspace (0,11, 1000)
pdf = chi2_rv.pdf(x)
plt.plot (x,pdf, label = 'Theoretical Chi2')

# Построение гистограммы
chi2_rvs = chi2_rv.rvs (1000)
plt.hist (chi2_rvs, density = True)

# Подпись осей
plt.ylabel ('$f(x)$')
plt.xlabel ('$x$')
plt.legend (loc = 'upper right')
```

## Работа с выборками

### Среднее и дисперсия

Мат. ожидание: $ E = k $

Дисперсия: $ D = 2k$

### Вычисление параметров нормальных распределений, аппроксимирующих выборочные средние
Для всех n: E(x) = 4 

Среднее отклонение $ \sigma = \sqrt{\frac{D(x)}{n}} = \sqrt{\frac{8}{n}}$

Таким образом
$$ \sigma_1 = 2.83 $$
$$ \sigma_5 = 1.26 $$
$$ \sigma_{10} = 0.9 $$
$$ \sigma_{25} = 0.57 $$
$$ \sigma_{50} = 0.4 $$

```python
D = 2*4
E = 4
```

### 1. 1000 выборок объема 1

```python
n = 1

# Генерация выборки
samples_of_1 = []
for i in range (1000):
    samples_of_1.append(chi2_rv.rvs(1))

# Подсчет среднего каждой выборки
sample_1_averages = []
for sample in samples_of_1:
    sample_1_averages.append(sample.mean())

# Построение гистограммы для выборочных средних
plt.hist (sample_1_averages, bins = 15, density = True)

# Распределение выборочного среднего при заданном параметре n
sample_1_average_rv = sts.norm (loc = E, scale = ((float(D)/float(n))**(0.5)))
x = np.linspace (0,18, 1000)
pdf = sample_1_average_rv.pdf(x)
plt.plot (x, pdf)

# Подпись оси
plt.xlabel ('$\\bar x$')
```
